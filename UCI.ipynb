
### URL and Domain Characteristics
1. FILENAME: The name of the file containing the URL data.
2. URL: The full URL string being analyzed.
3. URLLength: The total length of the URL; phishing URLs are often longer than legitimate ones.
4. Domain: The domain name of the URL.
5. DomainLength: The length of the domain; unusual domain lengths can indicate phishing.
6. IsDomainIP: A binary indicator of whether the domain is an IP address; phishing sites often use IP addresses directly.
7. TLD: The Top-Level Domain (e.g., .com, .net); some TLDs are more associated with phishing.
8. TLDLength: The length of the TLD, as unusual or overly long TLDs can sometimes correlate with phishing.
### URL Patterns and Probabilities
9. URLSimilarityIndex: A measure of how similar the URL is to legitimate ones, often based on string-matching algorithms.
10. CharContinuationRate: Measures the continuation of certain characters in the URL, which might suggest attempts to evade detection.
11. TLDLegitimateProb: The probability that the TLD is legitimate, based on pre-existing data.
12. URLCharProb: Probability of the characters used in the URL, indicating common phishing patterns.
13. NoOfSubDomain: Counts subdomains, as phishing URLs sometimes use extra subdomains to mimic legitimacy.
### Obfuscation Features
14. HasObfuscation: Indicates if the URL uses obfuscation, such as encoded characters or unusual patterns.
15. NoOfObfuscatedChar: Counts obfuscated characters within the URL.
16. ObfuscationRatio: The ratio of obfuscated characters to the total length of the URL.
### Character Analysis in URL
17. NoOfLettersInURL: The total number of letters within the URL.
18. LetterRatioInURL: Ratio of letters to the total length of the URL.
19. NoOfDegitsInURL: The count of digits in the URL, as phishing URLs may include numbers to mimic legitimacy.
20. DegitRatioInURL: Ratio of digits in the URL.
### Special Characters in URL
21. NoOfEqualsInURL: The count of = symbols, as they may be used in query strings.
22. NoOfQMarkInURL: Count of ? symbols, often used in URL parameters.
23. NoOfAmpersandInURL: Count of & symbols, common in complex query strings.
24. NoOfOtherSpecialCharsInURL: Counts other special characters in the URL.
25. SpacialCharRatioInURL: Ratio of special characters to the total length of the URL.
### Protocol and HTTP Information
26. IsHTTPS: A binary indicator of HTTPS usage. While not definitive, many phishing sites use HTTP.
27. LineOfCode: Number of lines of code in the page’s HTML.
28. LargestLineLength: Length of the longest line in the HTML, as overly lengthy lines could indicate suspicious embedded content.
### Page Content and Title Matching
29. HasTitle: Whether the HTML page has a title.
30. Title: The actual title of the HTML page.
31. DomainTitleMatchScore: The degree of similarity between the domain and page title.
32. URLTitleMatchScore: Similarity score between the URL and page title.
33. Security and Meta Elements
34. HasFavicon: Indicates if the site has a favicon (a small icon), which is common for legitimate sites.
35. Robots: Contents of the robots.txt file, which might indicate attempts to hide content from search engines.
36. IsResponsive: Checks if the site’s design is responsive, which can indicate a more legitimate setup.
### Redirects and Pop-ups
37. NoOfURLRedirect: Counts external redirects within the site.
38. NoOfSelfRedirect: Counts self-references that may be used to keep a visitor on the site.
39. NoOfPopup: Counts the number of pop-ups, as phishing sites may use pop-ups aggressively.
40. NoOfiFrame: Counts iframes, which can embed external content or redirect users.
### Form and Input Elements
41. HasExternalFormSubmit: Whether the form submission goes to an external domain, often a red flag.
42. HasSocialNet: Checks if the page has links to social networks, which can establish credibility.
43. HasSubmitButton: Indicates if the page contains a submit button, common in phishing sites.
44. HasHiddenFields: Hidden form fields can be used to capture information surreptitiously.
45. HasPasswordField: Indicates if the page contains a password field.
### Targeted Terms and Keywords
46. Bank: Indicates if banking terms appear in the content.
47. Pay: Indicates if the page mentions payment terms.
48. Crypto: Checks if cryptocurrency-related terms are present.
### Site Content Analysis
49. HasCopyrightInfo: Indicates if the site displays copyright information.
50. NoOfImage: Number of images on the page.
51. NoOfCSS: Number of CSS files loaded.
52. NoOfJS: Number of JavaScript files, as excessive scripts may indicate suspicious behavior.
Reference Types and Empty Links
53. NoOfSelfRef: Number of internal links, which may keep users on the phishing site.
54. NoOfEmptyRef: Count of empty or broken links, a potential sign of a poorly maintained phishing site.
55. NoOfExternalRef: Counts links to external sites, which could indicate credibility.
#### Label
56. label: The target variable indicating whether the URL is legitimate or phishing.
