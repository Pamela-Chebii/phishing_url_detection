### Working with both datasets

- Dataset1 and Dataset2 are both phishing datasets. We therefore need to identify similar columns as well as unique columns before combining the two datasets

## **`Common or Similar Columns`**
### **`URL Details`**

- url (Dataset 1) ↔️ URL (Dataset 2)
- length_url (Dataset 1) ↔️ URLLength (Dataset 2)
- length_hostname (Dataset 1) ↔️ DomainLength (Dataset 2)
- ip (Dataset 1) ↔️ IsDomainIP (Dataset 2) – boolean indicators of IP addresses.

### **`Special Characters`**

- Dataset 1 has detailed counts of various characters (e.g., nb_qm, nb_and, etc.) which correspond to general character counts in Dataset 2:

- nb_qm ↔️ NoOfQMarkInURL
- nb_and ↔️ NoOfAmpersandInURL
- nb_eq ↔️ NoOfEqualsInURL
- nb_dots ↔️ part of general NoOfOtherSpecialCharsInURL
- ratio_digits_url (Dataset 1) ↔️ DegitRatioInURL (Dataset 2)
- ratio_digits_host (Dataset 1) ↔️ LetterRatioInURL (Dataset 2)

### **`Subdomains`**
- nb_subdomains (Dataset 1) ↔️ NoOfSubDomain (Dataset 2)

**`HTTPS and Security`**
- https_token (Dataset 1) ↔️ IsHTTPS (Dataset 2)

### **`Domain and Brand Matches`**
- `domain_in_brand`, `brand_in_subdomain`, `brand_in_path` (Dataset 1) ↔️ `DomainTitleMatchScore` and `URLTitleMatchScore` (Dataset 2) – both reflect domain and brand associations.

**`HTML Elements & Embedded Content`**
- external_favicon (Dataset 1) ↔️ HasFavicon (Dataset 2)
- links_in_tags (Dataset 1) ↔️ NoOfExternalRef and NoOfSelfRef (Dataset 2)
- submit_email (Dataset 1) ↔️ HasSubmitButton (Dataset 2)
-iframe (Dataset 1) ↔️ NoOfiFrame (Dataset 2)
- popup_window (Dataset 1) ↔️ NoOfPopup (Dataset 2)
- domain_with_copyright (Dataset 1) ↔️ HasCopyrightInfo (Dataset 2)

### **`Miscellaneous Indicators`**
- web_traffic (Dataset 1) ↔️ Similar to Robots and IsResponsive in Dataset 2 – these columns may represent engagement or activity level indicators.

### **`Unique Columns in Each Dataset`**

#### **Dataset 1 Specific Columns:**

- Many columns are detailed counts of various special characters (e.g., `nb_at`, `nb_hyphens`, `nb_tilde`, etc.)
- Additional technical indicators like `port`, `tld_in_path`, and `statistical_report`
- `Whois` and registration details like `domain_age` and `domain_registration_length`

#### **Dataset 2 Specific Columns:**
- Contains unique columns related to the web structure, like `NoOfImage`, `NoOfCSS`, `NoOfJS`, `Bank`, `Pay`, `Crypto`, and `HasSocialNet`

## MERGING DATASET 1 AND DATASET 2

The datasets will be merged as follows:
1. Standardize Column Names: Rename similar columns in each dataset to a common name.
2. Align Columns: Add missing columns in each dataset as NaN.
3. Concatenate: Use pd.concat to combine the datasets.

```python
# Standardize similar column names
Dataset1.rename(columns={
    'url': 'URL', 'length_url': 'URLLength', 'length_hostname': 'DomainLength',
    'ip': 'IsDomainIP', 'nb_qm': 'NoOfQMarkInURL', 'nb_and': 'NoOfAmpersandInURL',
    'nb_eq': 'NoOfEqualsInURL', 'nb_subdomains': 'NoOfSubDomain',
    'https_token': 'IsHTTPS', 'external_favicon': 'HasFavicon', 
    'submit_email': 'HasSubmitButton', 'iframe': 'NoOfiFrame', 
    'popup_window': 'NoOfPopup', 'domain_with_copyright': 'HasCopyrightInfo', 'status': 'label'
    # Add other similar mappings as needed
}, inplace=True)

# Add missing columns to each dataset
for column in set(Dataset1.columns).difference(Dataset2.columns):
    Dataset2[column] = pd.NA
for column in set(Dataset2.columns).difference(Dataset1.columns):
    Dataset1[column] = pd.NA

# Combine the datasets
combined_data = pd.concat([Dataset1, Dataset2], ignore_index=True)

```
